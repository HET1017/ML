```python
# Separate the data, test and train
X=data.drop('liked',axis=1)

y=data[['liked']]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state= 50)
```

According to the features in the data, I wanted to make a prediction about whether the music would be liked or not, and I used the classification method because it was a definite result. liked or not(0-1).

**Discussion**

Since the output of the liked parameter is 0-1, I used the classification method. For this, first I used logistic regression, then desicion tree and finally random forest.
The classification algorithm that gave the best results in the models I made was random forest.
The values of logistic regression and desicion tree algorithms were close to each other. dct is better if you need to make a comparison.
However, random forest outperformed both algorithms with a f1 score of 92%.

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score ,classification_report

model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Print the performance output
print(classification_report(y_test, predictions))
print('Predicted labels: ', predictions)
print('Accuracy: ', accuracy_score(y_test, predictions))
print('Precision:', precision_score(y_test, predictions))
print('Recall:', recall_score(y_test, predictions))
print('F1 score:', f1_score(y_test, predictions))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(random_state=77, max_depth=5)
DTC.fit(X_train, y_train)

# Make predictions
predictions = DTC.predict(X_test)

# Print the performance output
print(classification_report(y_test, predictions))
print('Predicted labels: ', predictions)
print('Accuracy:', accuracy_score(y_test, predictions))
print('Precision:', precision_score(y_test, predictions))
print('Recall:', recall_score(y_test, predictions))
print('F1 score:', f1_score(y_test, predictions))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state=77, n_estimators=200, max_depth=10)
RF.fit(X_train, y_train)

# Make predictions
predictions = RF.predict(X_test)

# Print the performance output
print(classification_report(y_test, predictions))
print('Predicted labels: ', predictions)
print('Accuracy:', accuracy_score(y_test, predictions))
print('Precision:', precision_score(y_test, predictions))
print('Recall:', recall_score(y_test, predictions))
print('F1 score:', f1_score(y_test, predictions))
```
